```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
data = {
    'Number UEs': ['16', '32', '64', '128', '256', '512', '1024'],
    'E2Nodes ON': [1, 1, 1, 2, 5, 9, 17],
    'E2Nodes OFF':[16, 16, 16, 15, 12, 8, 0],
    'Energy Consumed (W)' : [15.4757, 15.4757, 15.4757, 30.9521, 77.3785, 139.2813, 263.0869]
}

df = pd.DataFrame(data)
```

```python
# Set seaborn style
sns.set(style="darkgrid")

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot E2Nodes ON and OFF
ax1.plot(df['Number UEs'], df['E2Nodes ON'], marker='o', label='E2Nodes ON', color='tab:blue')
ax1.plot(df['Number UEs'], df['E2Nodes OFF'], marker='o', label='E2Nodes OFF', color='tab:grey')

# Add titles and labels to the first y-axis
ax1.set_title('E2Nodes ON/OFF and Energy Consumed vs Number of UEs')
ax1.set_xlabel('Number of UEs')
ax1.set_ylabel('Number of E2Nodes')
ax1.legend(loc='center left')

# Annotate the values for E2Nodes ON and OFF
for i in range(len(df)):
    ax1.annotate(df['E2Nodes ON'][i], (df['Number UEs'][i], df['E2Nodes ON'][i]), textcoords="offset points", xytext=(0,10), ha='center', color='tab:blue')
    ax1.annotate(df['E2Nodes OFF'][i], (df['Number UEs'][i], df['E2Nodes OFF'][i]), textcoords="offset points", xytext=(0,10), ha='center', color='tab:grey')

# Create a secondary y-axis for Energy Consumed
ax2 = ax1.twinx()
ax2.plot(df['Number UEs'], df['Energy Consumed (W)'], marker='o', label='Energy Consumed (W)', color='tab:red')

# Add labels to the secondary y-axis
ax2.set_ylabel('Energy Consumed (W)', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Annotate the values for Energy Consumed
for i in range(len(df)):
    ax2.annotate(round(df['Energy Consumed (W)'][i], 2), (df['Number UEs'][i], df['Energy Consumed (W)'][i]), textcoords="offset points", xytext=(20, -20), ha='center', color='tab:red')

# Add legend for the secondary y-axis
ax2.legend(loc='lower right')

# Show the plot
plt.tight_layout()
plt.show()
```

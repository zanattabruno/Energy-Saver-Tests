```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Data
data = {
    'Number UEs': ['16', '32', '64', '128', '256', '512', '1024'],
    'E2Nodes ON': [1, 1, 1, 2, 5, 9, 17],
    'E2Nodes OFF':[16, 16, 16, 15, 12, 8, 0],
    'Max CPU (millicores)': [0.0881, 0.013, 0.477, 0.938, 1.9, 3.8, 7.6],
    'Std Dev CPU': [0.011, 0.002, 0.096, 0.14, 0.38, 0.57, 1.52],
    'Max Memory (MB)': [45.5, 58.1, 107.83, 214.7, 366.77, 626.59, 1070.58],
    'Std Dev Memory': [3.2, 6, 8.5, 15, 25, 50, 90],
    'Time to solve (s)': [1.56, 2.55, 6.42, 20.24, 49.26, 119.95, 292.04],
    'Std Dev Time': [0.39, 0.38, 1.29, 3.24, 8.37, 14.39, 73.01]
}

df = pd.DataFrame(data)

# Set seaborn style
sns.set(style="darkgrid")
```

```python
# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10), sharex=True)

# Stacked bar plot for E2nodes ON and OFF vs Number UEs
#width = 0.4  # the width of the bars
axs[0, 0].bar(df['Number UEs'], df['E2Nodes ON'], width, label='E2Nodes ON', color='tab:blue')
axs[0, 0].bar(df['Number UEs'], df['E2Nodes OFF'], width, bottom=df['E2Nodes ON'], label='E2Nodes OFF', color='tab:grey')

axs[0, 0].set_title('Number of E2Nodes ON and OFF vs Number of UEs')
axs[0, 0].set_xlabel('Number of UEs')
axs[0, 0].set_ylabel('Number of E2Nodes')
axs[0, 0].legend()

# Line plot for Max CPU vs Number UEs
sns.barplot(x='Number UEs', y='Max CPU (millicores)', yerr=df['Std Dev CPU'], data=df, ax=axs[0, 1], color='tab:orange')
axs[0, 1].set_title('Max CPU vs Number of UEs')
axs[0, 1].set_xlabel('Number of UEs')
axs[0, 1].set_ylabel('Max CPU (millicores)')

# Line plot for Max Memory vs Number UEs
sns.barplot(x='Number UEs', y='Max Memory (MB)', yerr=df['Std Dev Memory'], data=df, ax=axs[1, 0], color='tab:green')
axs[1, 0].set_title('Max Memory vs Number of UEs')
axs[1, 0].set_xlabel('Number of UEs')
axs[1, 0].set_ylabel('Max Memory (MB)')

# Line plot for Time to solve vs Number UEs
sns.barplot(x='Number UEs', y='Time to solve (s)', yerr=df['Std Dev Time'], data=df, ax=axs[1, 1], color='tab:red')
axs[1, 1].set_title('Time to Solve vs Number of UEs')
axs[1, 1].set_xlabel('Number of UEs')
axs[1, 1].set_ylabel('Time to solve (s)')

# Adjust layout
plt.tight_layout()
```

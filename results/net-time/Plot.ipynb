```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#data
data = {
    'Number UEs': [16, 32, 64, 128, 256, 512, 1024],
    'Gross time end-to-end (s)': [69, 77, 89, 122, 179.3, 308.2, 665.7],
    'Gross Std Dev (s)': [21.2, 12.7, 29.3, 48.5, 59.3, 68.2, 185.7],
    'Wait to start (s)': [60, 60, 60, 90, 120, 240, 480],
    'Net time end-to-end (s)': [9, 17, 29, 32, 59.3, 68.2, 185.7]
}
df = pd.DataFrame(data)

# Set seaborn style
sns.set(style="darkgrid")
```

```python
# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

# Bar plot for Gross time end-to-end vs Number UEs
sns.barplot(x='Number UEs', y='Gross time end-to-end (s)', yerr=df['Gross Std Dev (s)'], data=df, ax=axs[0], color='tab:blue')
axs[0].set_title('Gross Time End-to-End vs Number of UEs')
axs[0].set_ylabel('Gross Time End-to-End (s)')

# Annotate the values of each bar
for i, v in enumerate(df['Gross time end-to-end (s)']):
    axs[0].annotate(str(round(v, 1)), (i, v + df['Gross Std Dev (s)'][i] + 1), ha='center', va='bottom')

# Bar plot for Wait to Start vs Number UEs
sns.barplot(x='Number UEs', y='Wait to start (s)', data=df, ax=axs[1], color='tab:green')
axs[1].set_title('Wait to Start vs Number of UEs')
axs[1].set_ylabel('Wait to Start (s)')

# Annotate the values of each bar
for i, v in enumerate(df['Wait to start (s)']):
    axs[1].annotate(str(round(v, 1)), (i, v + 1), ha='center', va='bottom')

# Bar plot for Net time end-to-end vs Number UEs
sns.barplot(x='Number UEs', y='Net time end-to-end (s)', data=df, ax=axs[2], color='tab:red')
axs[2].set_title('Net Time End-to-End vs Number of UEs')
axs[2].set_xlabel('Number of UEs')
axs[2].set_ylabel('Net Time End-to-End (s)')

# Annotate the values of each bar
for i, v in enumerate(df['Net time end-to-end (s)']):
    axs[2].annotate(str(round(v, 1)), (i, v + 1), ha='center', va='bottom')

# Adjust layout
plt.tight_layout()
plt.savefig('out/end-to-end.pdf', bbox_inches='tight')
plt.savefig('out/end-to-end.png', dpi=300, bbox_inches='tight')
plt.show()
```

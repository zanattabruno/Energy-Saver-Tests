```python
import numpy as np
import matplotlib.pyplot as plt

# Time parameters (in minutes)
T_ramp = 60    # Duration of entrance phase (ramp-up)
T_game = 120   # Duration of the game (peak phase)
T_exit = 60    # Duration of exit phase (ramp-down)

# Total event duration
T_total = T_ramp + T_game + T_exit

# Time array (in minutes)
t = np.linspace(0, T_total, 1000)

# Peak number of active users (for example, 30,000)
N_peak = 30000

# Define a piecewise function for active users over time:
def active_users(time):
    if time < T_ramp:
        # Entrance phase: linear increase
        return N_peak * (time / T_ramp)
    elif time < T_ramp + T_game:
        # Game phase: constant peak load
        return N_peak
    else:
        # Exit phase: linear decrease
        return N_peak * (1 - (time - (T_ramp + T_game)) / T_exit)

# Vectorize the function so it can be applied element-wise
active_users_vec = np.vectorize(active_users)
N_t = active_users_vec(t)

# Plot the active users over time
plt.figure(figsize=(10, 6))
plt.plot(t, N_t, linewidth=2, color='blue')
plt.xlabel("Time (minutes)")
plt.ylabel("Number of Active Users")
plt.title("Simulation of Active Users Over a Stadium Event")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define event duration parameters (in minutes)
T_ramp = 60    # Entrance phase duration (ramp-up)
T_game = 120   # Game phase duration (peak)
T_exit = 60    # Exit phase duration (ramp-down)
T_total = T_ramp + T_game + T_exit

# Number of time points for the simulation
n_points = 1000
time = np.linspace(0, T_total, n_points)

# Define the peak number of active users
N_peak = 30000

# Create a deterministic baseline for active users using a piecewise model
def active_users_deterministic(t):
    if t < T_ramp:
        # Ramp-up phase: linear increase
        return N_peak * (t / T_ramp)
    elif t < T_ramp + T_game:
        # Peak phase: constant number of users
        return N_peak
    else:
        # Ramp-down phase: linear decrease
        return N_peak * (1 - (t - (T_ramp + T_game)) / T_exit)

# Vectorize the function for the entire time array
active_users_det = np.array([active_users_deterministic(ti) for ti in time])

# Add aleatoriety: for instance, add Gaussian noise (10% of peak level)
noise_std = 0.1 * N_peak  # standard deviation as 10% of N_peak
noise = np.random.normal(0, noise_std, n_points)

# Active users with noise (clip at 0 to avoid negative users)
active_users_noisy = active_users_det + noise
active_users_noisy = np.clip(active_users_noisy, 0, None)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(time, active_users_noisy, label="Active Users (with noise)", color="blue")
plt.plot(time, active_users_det, label="Deterministic Baseline", color="red", linestyle="--")
plt.xlabel("Time (minutes)")
plt.ylabel("Number of Active Users")
plt.title("Simulated Active Users Over a Stadium Event")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1) Define the 24-hour timeline
# -----------------------------
hours = np.arange(0, 24.01, 0.25)  # every 15 minutes

# -------------------------------------------------
# 2) Create a "weekday baseline" without an event
#    (just a smooth daily cycle + random variation)
# -------------------------------------------------
# We can approximate a "daily cycle" with a sinusoid
# that peaks around midday or mid-afternoon.
weekday_baseline = 500 + 300 * np.sin((hours - 8) * np.pi / 12)
# Add some random noise
np.random.seed(42)
weekday_baseline += np.random.normal(0, 40, len(weekday_baseline))
# Ensure non-negative
weekday_baseline = np.clip(weekday_baseline, 0, None)

# ------------------------------------------------
# 3) Create an "event day" curve with a big match
#    - ramp-up before the game
#    - peak around break/end
#    - ramp-down afterwards
# ------------------------------------------------
event_curve = np.zeros_like(hours)

# Example schedule:
#   - People start arriving heavily from ~16:00 to 19:30
#   - Game runs ~19:30 to 21:15, with a halftime ~20:15
#   - Big exit peak after 21:15 until ~22:30
#   - Then drop

# Let's define piecewise segments:
for i, h in enumerate(hours):
    if h < 16:
        # normal traffic until 16:00
        event_curve[i] = 500 + 100 * np.sin((h - 8) * np.pi / 12)
    elif 16 <= h < 19.5:
        # ramp up from 16:00 to 19:30
        # from ~1000 up to ~5000
        fraction = (h - 16) / (19.5 - 16)
        event_curve[i] = 1000 + fraction * (5000 - 1000)
    elif 19.5 <= h < 20.25:
        # short plateau near start of game
        event_curve[i] = 5000
    elif 20.25 <= h < 20.5:
        # short halftime peak (e.g. 7000)
        # from 20:15 to 20:30
        fraction = (h - 20.25) / (0.25)  # 15 minutes
        event_curve[i] = 5000 + fraction * (7000 - 5000)
    elif 20.5 <= h < 21.25:
        # after halftime, still high
        event_curve[i] = 7000
    elif 21.25 <= h < 22.5:
        # big exit peak from 21:15 to 22:30
        fraction = (h - 21.25) / (1.25)
        event_curve[i] = 7000 + fraction * (8000 - 7000)
    else:
        # drop after 22:30
        fraction = min((h - 22.5) / (24 - 22.5), 1.0)
        event_curve[i] = 8000 * max(0, 1 - fraction)

# Add random fluctuations to the event curve
np.random.seed(123)
event_curve += np.random.normal(0, 200, len(event_curve))
event_curve = np.clip(event_curve, 0, None)

# ---------------------------------------
# 4) Compute peak-to-average ratio (PAR)
# ---------------------------------------
def peak_to_avg_ratio(curve, window=1):
    """
    Roughly compute a rolling peak-to-average ratio over a window in hours.
    For simplicity, we just do local max / local mean in a small range.
    """
    n_window = int(window / 0.25)  # how many points in "window" hours
    par_values = []
    for i in range(len(curve)):
        start = max(i - n_window // 2, 0)
        end = min(i + n_window // 2, len(curve))
        segment = curve[start:end]
        if segment.mean() > 0:
            par_values.append(segment.max() / segment.mean())
        else:
            par_values.append(1.0)
    return np.array(par_values)

weekday_par = peak_to_avg_ratio(weekday_baseline, window=1)
event_par = peak_to_avg_ratio(event_curve, window=1)

# --------------
# 5) Plot Results
# --------------
fig, ax1 = plt.subplots(figsize=(10,6))

# Plot RRC connected devices (or user count)
ax1.plot(hours, weekday_baseline, '--k', label='Weekday Baseline (avg)', linewidth=2)
ax1.plot(hours, event_curve, '-b', label='Event Day (avg)', linewidth=2)
ax1.set_xlabel('Time (hours)')
ax1.set_ylabel('RRC Connected Devices (per 15 minutes)', color='k')
ax1.set_ylim(0, 9000)

# Plot the peak-to-avg ratio on a second y-axis
ax2 = ax1.twinx()
ax2.plot(hours, weekday_par, '--', color='gray', label='Weekday PAR')
ax2.plot(hours, event_par, '-.', color='red', label='Event PAR')
ax2.set_ylabel('Peak-to-Average Ratio', color='gray')
ax2.set_ylim(0, 3)

# Add vertical lines for game start/end (approx)
ax1.axvline(x=19.5, color='gray', linestyle=':', linewidth=1)
ax1.axvline(x=21.25, color='gray', linestyle=':', linewidth=1)
ax1.fill_betweenx([0,9000],19.5,21.25,color='red',alpha=0.05,label='Game Window')

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Example: Event Day vs. Weekday, with Peak-to-Average Ratio')
plt.grid(True)
plt.show()
```
